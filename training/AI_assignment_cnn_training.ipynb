**Task 2:Model building**

**Deep learning-CNN**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model=Sequential()

model.add(Conv2D(32,(3,3),activation="relu",input_shape=(64,64,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.summary()
```

```python
import tensorflow as tf

IMG_SIZE =(64,64)
BATCH_SIZE =32
train_dataset=tf.keras.utils.image_dataset_from_directory("/content/drive/MyDrive/ml dataset/dataset-20250410T042537Z-001/dataset/training_set",image_size=IMG_SIZE,batch_size=BATCH_SIZE)
test_dataset=tf.keras.utils.image_dataset_from_directory("/content/drive/MyDrive/ml dataset/dataset-20250410T042537Z-001/dataset/test_set",image_size=IMG_SIZE,batch_size=BATCH_SIZE)
```

```python
model.fit(train_dataset, epochs=10, validation_data=test_dataset)

```

1.why this model was choosen?

images have spatial structure.
CNNs automatically extract features like edges, shapes, and textures.
fewer parameters than fully connected networks.
high performance in image classification tasks.

2.How features impact prediction?

early layers detect edges,corners.
middle layers used for shapes,textures.
deeper layers uded for object specific patterns(ears, faces).
good features causes higher accuracy.
poor features causes for misclassification.

3.what improvements can be done?

add more Conv layers.
increase filters.
use Batch Normalization.
add Dropout to reduce overfitting.
